```python
from arco.cli.commands.bench import load_benchmark_dataset
bench_dataset = load_benchmark_dataset(dataset)
def show_diff(gt_df, benchmark_df, idx):
    prompt = gt_df[idx]["prompt"]
    analysis = benchmark_df["Analyzer_1_message"].loc[idx]
    gt_analysis = gt_df[idx]["gt_analysis"]
    perplexity = benchmark_df["Analyzer_1_perplexity"].loc[idx]
    accuracy = benchmark_df["Analyzer_1_evaluation_gt"].loc[idx]
    print(f"""
    Prompt: {prompt}
    Analysis: {analysis}
    GT Analysis: {gt_analysis}
    Perplexity: {perplexity}
    Accuracy: {accuracy}
    """)
```

```python
import pandas as pd
from scipy.stats import pearsonr

def compute_result_correlation(df):
    columns = ["Retriever_1", "Analyzer_1", "Visualizer_1"]
    results = {}

    all_gt = []
    all_perplexity = []

    for column in columns:
        gt = df[f"{column}_evaluation_gt"]
        perp = df[f"{column}_perplexity"]

        # Remove missing values
        valid = pd.concat([gt, perp], axis=1).dropna()
        gt = valid.iloc[:, 0]
        perp = valid.iloc[:, 1]

        if len(gt) < 2 or gt.nunique() <= 1 or perp.nunique() <= 1:
            if len(gt) < 2 : print(f"len(gt)={len(gt)}")
            if gt.nunique() <=1 : print(f"gt.nunique() = {gt.nunique()}")
            if perp.nunique() <=1 : print(f"perp.nunique() = {perp.nunique()}")
            corr, p_value = float("nan"), float("nan")
        else:
            corr, p_value = pearsonr(gt, perp)

        results[column] = {"correlation": corr, "p_value": p_value}

        all_gt.extend(gt)
        all_perplexity.extend(perp)

    all_gt = pd.Series(all_gt)
    all_perplexity = pd.Series(all_perplexity)

    if all_gt.nunique() <= 1 or all_perplexity.nunique() <= 1:
        overall_corr, overall_p = float("nan"), float("nan")
    else:
        overall_corr, overall_p = pearsonr(all_gt, all_perplexity)

    print("Per-column correlations:")
    for col, res in results.items():
        print(
            f"  {col}: r = {res['correlation']:.4f}, "
            f"p = {res['p_value']:.4f}"
        )

    print(f"\nOverall: r = {overall_corr:.4f}, p = {overall_p:.4f}")

    return results, (overall_corr, overall_p)
```

```python
corr_low, overall_corr_low = compute_result_correlation(low_temp_df)
```

```python
corr_high, overall_corr_high = compute_result_correlation(high_temp_df)
```
