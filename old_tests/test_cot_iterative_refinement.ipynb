### Shared test helpers

```python
class FakeLLM:
    """Mock LLM: returns a fixed response and records all prompts received."""
    def __init__(self, response_text=''):
        self._response = response_text
        self.calls = []

    def invoke(self, prompt):
        self.calls.append(prompt)
        class _R:
            pass
        r = _R()
        r.content = self._response
        return r

    def __getattr__(self, name):
        return None


def make_core_fn(responses):
    """Returns a core_fn that steps through `responses` on each call."""
    idx = [0]
    call_log = []
    def core_fn(state, llm):
        sql = responses[min(idx[0], len(responses) - 1)]
        idx[0] += 1
        call_log.append(sql)
        return {'sql_query': sql}
    core_fn.call_log = call_log
    return core_fn


# Shared agent instance (no LLM calls during __init__)
agent = SalesDataWorkflow(model='gpt-4o-mini')
dummy_state = {'prompt': 'test'}
dummy_llm   = FakeLLM()

print('Helpers ready.')
```

## 2. Unit Tests — CoT Helpers (no API key required)

```python
print('=== Test 2.1: _extract_step_output ===')

assert _extract_step_output('lookup_sales_data', {'sql_query': 'SELECT 1'}) == 'SELECT 1'
assert _extract_step_output('decide_tool', {'tool_choice': 'analyzing_data'}) == 'analyzing_data'
assert _extract_step_output('analyzing_data', {'answer': ['first', 'second']}) == 'second'
assert _extract_step_output('create_visualization', {'answer': ['chart_code']}) == 'chart_code'
assert _extract_step_output('analyzing_data', {}) == ''  # empty fallback

print('  All assertions passed ✓')
```

```python
print('=== Test 2.2: CoTRefinementLLM prompt augmentation ===')

base = FakeLLM('SELECT 2 FROM sales')
wrapper = CoTRefiner(base, previous_response='SELECT 1 FROM sales')
wrapper.invoke('Original prompt here')

received = base.calls[0]
assert 'Original prompt here' in received, 'Original prompt must be present'
assert 'ITERATIVE REFINEMENT'  in received, 'Refinement header must be present'
assert 'SELECT 1 FROM sales'   in received, 'Previous response must be quoted'

print('  Refinement suffix correctly appended ✓')
print(f'  Prompt grew from {len("Original prompt here")} to {len(received)} chars')
```

```python
print('=== Test 2.3: early stop on identical output ===')

# Iteration 2 returns same SQL → early stop; iteration 3 never executed
fn = make_core_fn(['SELECT SUM(a) FROM t', 'SELECT AVG(a) FROM t'])
initial = {'sql_query': 'SELECT SUM(a) FROM t'}

result = agent._apply_cot_iterations(
    step_name='lookup_sales_data', state=dummy_state,
    core_fn=fn, llm=dummy_llm, initial_result=initial, cot_n=3,
)

assert result['sql_query'] == 'SELECT SUM(a) FROM t', f"Unexpected: {result['sql_query']}"
assert len(fn.call_log) == 1, f'Expected 1 refinement call (early stop), got {len(fn.call_log)}'

print(f'  Early stop after 1 refinement call ✓  (3rd iteration not reached)')
```

```python
print('=== Test 2.4: all cot_n iterations run when outputs are dissimilar ===')
print(f'  similarity threshold = {_COT_SIMILARITY_THRESHOLD}')

# SELECT / UPDATE / DELETE are structurally incompatible → similarity << 0.95
s1 = 'SELECT SUM(revenue) FROM sales'
s2 = 'UPDATE inventory SET quantity = quantity - 1 WHERE product_id = 42'
s3 = 'DELETE FROM expired_sessions WHERE last_active < NOW() - INTERVAL 30 days'

r12 = difflib.SequenceMatcher(None, s1, s2).ratio()
r23 = difflib.SequenceMatcher(None, s2, s3).ratio()
print(f'  similarity(s1,s2)={r12:.3f}   similarity(s2,s3)={r23:.3f}')
assert r12 < _COT_SIMILARITY_THRESHOLD, f'Test strings too similar: {r12:.3f}'
assert r23 < _COT_SIMILARITY_THRESHOLD, f'Test strings too similar: {r23:.3f}'

fn2 = make_core_fn([s2, s3])
result2 = agent._apply_cot_iterations(
    step_name='lookup_sales_data', state=dummy_state,
    core_fn=fn2, llm=dummy_llm, initial_result={'sql_query': s1}, cot_n=3,
)

assert result2['sql_query'] == s3, f"Unexpected final SQL: {result2['sql_query']}"
assert len(fn2.call_log) == 2, f'Expected 2 refinement calls, got {len(fn2.call_log)}'

print(f'  All {len(fn2.call_log)} refinement iterations executed ✓')
```

```python
print('=== Test 2.5: cot_n=1 is a strict no-op ===')

fn3 = make_core_fn(['SHOULD NEVER BE CALLED'])
initial3 = {'sql_query': 'SELECT 1'}

result3 = agent._apply_cot_iterations(
    step_name='lookup_sales_data', state=dummy_state,
    core_fn=fn3, llm=dummy_llm, initial_result=initial3, cot_n=1,
)

assert result3 is initial3, 'cot_n=1 must return the exact same object'
assert len(fn3.call_log) == 0, 'cot_n=1 must make zero extra LLM calls'

print('  cot_n=1: initial result returned unchanged, zero extra calls ✓')
print('\n=== All unit tests passed ✓ ===')
```

## 7. Unit Tests — Python Code Validation in create_visualization_core

These tests verify that `create_visualization_core` now **executes** the generated
matplotlib code headlessly and surfaces any runtime error in the result's `"error"` key,
so the CoT refinement loop can feed it back to the LLM.

**No API key required** — a `MultiResponseFakeLLM` drives both LLM calls
(chart-config extraction and code generation).

| # | What is tested |
|---|----------------|
| 7.1 | Valid code → result has **no** `"error"` key |
| 7.2 | Broken code (bad column name) → `"error"` key contains the `KeyError` |
| 7.3 | `_apply_cot_iterations` for `create_visualization` with error on iteration 1, success on iteration 2 → final result clean and error was forwarded to the LLM |

```python
# --- Shared helpers for Section 7 ---

import pandas as pd

class MultiResponseFakeLLM:
    """FakeLLM that steps through a list of fixed responses, one per invoke() call."""
    def __init__(self, responses):
        self._iter = iter(responses)
        self.calls = []

    def invoke(self, prompt):
        self.calls.append(prompt)
        text = next(self._iter)
        return type('_R', (), {'content': text})()

    def __getattr__(self, name):
        return None


# Minimal DataFrame used by all tests in this section
_test_df = pd.DataFrame({
    'month':   ['Jan', 'Feb', 'Mar'],
    'revenue': [100,   200,   150],
})

# Reusable JSON config response (valid for _test_df)
_VALID_CONFIG_JSON = '{"chart_type": "bar", "x_axis": "month", "y_axis": "revenue", "title": "Revenue by Month"}'

# Valid matplotlib code that works with _test_df and the config above
_VALID_CODE = """\
import matplotlib.pyplot as plt
import pandas as pd
x_data = data_df[config['x_axis']]
y_data = data_df[config['y_axis']]
plt.figure(figsize=(8, 4))
plt.bar(x_data, y_data)
plt.xlabel(config['x_axis'])
plt.ylabel(config['y_axis'])
plt.title(config['title'])
plt.tight_layout()
plt.show()
"""

# Broken code: references a column that does not exist → KeyError at exec time
_BROKEN_CODE = """\
import matplotlib.pyplot as plt
import pandas as pd
x_data = data_df[config['x_axis']]
y_data = data_df['nonexistent_column']  # KeyError!
plt.bar(x_data, y_data)
plt.show()
"""

print('Section 7 helpers ready.')
```

```python
print('=== Test 7.1: valid code → no error in result ===')

llm_71 = MultiResponseFakeLLM([_VALID_CONFIG_JSON, _VALID_CODE])
state_71 = {'prompt': 'Show revenue by month', 'data': 'month revenue\nJan 100', 'data_df': _test_df, 'answer': []}

result_71 = create_visualization_core(state_71, llm_71)

assert not result_71.get('error'), f"Unexpected error: {result_71.get('error')}"
assert result_71.get('answer'), "Expected generated code in 'answer'"
assert result_71['answer'][-1].strip() == _VALID_CODE.strip(), "Returned code should match generated code"
assert llm_71.calls[0], "First LLM call (config extraction) must have happened"
assert llm_71.calls[1], "Second LLM call (code generation) must have happened"

print(f'  chart_config: {result_71.get("chart_config")}')
print(f'  answer length: {len(result_71["answer"][-1])} chars')
print('  No error in result ✓')
print('  Test 7.1 passed ✓')
```

```python
print('=== Test 7.2: broken code (bad column) → error captured in result ===')

llm_72 = MultiResponseFakeLLM([_VALID_CONFIG_JSON, _BROKEN_CODE])
state_72 = {'prompt': 'Show revenue by month', 'data': 'month revenue\nJan 100', 'data_df': _test_df, 'answer': []}

result_72 = create_visualization_core(state_72, llm_72)

exec_error = result_72.get('error', '')
assert exec_error, "Expected 'error' key in result when code raises at exec time"
assert 'KeyError' in exec_error, f"Expected KeyError in error message, got: {exec_error}"
assert 'nonexistent_column' in exec_error, f"Expected column name in error message, got: {exec_error}"
# The broken code is still stored in answer (so the LLM can see what it generated)
assert result_72.get('answer'), "Generated code should still be stored in 'answer' even on error"

print(f'  Captured error: {exec_error}')
print('  "error" key present and contains KeyError ✓')
print('  Generated code still stored in "answer" ✓')
print('  Test 7.2 passed ✓')
```

```python
print('=== Test 7.3: _apply_cot_iterations for create_visualization — error on iter 1, success on iter 2, validation on iter 3 ===')
print('    Verifies: error forwarded to LLM prompt, final result has no error\n')

# core_fn that returns broken code on the first call, valid code on the second
_call_counter_73 = [0]

def _vis_core_fn_73(state, llm):
    """Simulate create_visualization_core: error on call 1, success on call 2, validation on call 3."""
    _call_counter_73[0] += 1
    # Invoke the LLM so CoTRefinementLLM can inject the suffix; we capture the
    # augmented prompt via the base FakeLLM's call log.
    llm.invoke('dummy chart prompt')

    if _call_counter_73[0] == 1:
        return {
            **state,
            'answer': [_BROKEN_CODE],
            'error': "KeyError: 'nonexistent_column'",
        }
    else:
        return {
            **state,
            'answer': [_VALID_CODE],
            # no 'error' key → success
        }


base_llm_73 = FakeLLM('some response')

# Run the "initial" call (iteration 0) manually, simulating what the agent does
# before handing off to _apply_cot_iterations.
_call_counter_73[0] = 0
initial_result_73 = _vis_core_fn_73(
    {'prompt': 'test', 'data': '', 'data_df': _test_df, 'answer': []},
    FakeLLM('initial'),
)
print(f'  Initial result error: {initial_result_73.get("error")}')

# cot_n=2 → exactly one refinement slot: the loop runs range(1, 2) = [1] only.
# This means _vis_core_fn_73 is called once more (counter → 2), regardless of
# similarity, so the counter stays deterministically at 2.
final_result_73 = agent._apply_cot_iterations(
    step_name='create_visualization',
    state={'prompt': 'test', 'data': '', 'data_df': _test_df, 'answer': []},
    core_fn=_vis_core_fn_73,
    llm=base_llm_73,
    initial_result=initial_result_73,
    cot_n=4,
)

# --- Assertions ---
assert not final_result_73.get('error'), \
    f"Final result should have no error after recovery, got: {final_result_73.get('error')}"

assert _call_counter_73[0] == 3, \
    f"Expected exactly 2 core_fn calls (1 initial + 1 refinement), got {_call_counter_73[0]}"

# Check that the error was injected into the LLM prompt during the refinement call.
# CoTRefinementLLM uses _ERROR_SUFFIX when execution_error is non-empty, so
# base_llm_73.calls[0] should be 'dummy chart prompt' + _ERROR_SUFFIX content.
error_prompt = base_llm_73.calls[0] if base_llm_73.calls else ''
assert 'EXECUTION ERROR' in error_prompt, \
    f"Expected _ERROR_SUFFIX in refinement prompt, got:\n{error_prompt[:300]}"
assert "KeyError: 'nonexistent_column'" in error_prompt, \
    "Error message must appear verbatim in the refinement prompt"

print(f'  core_fn called {_call_counter_73[0]} times (1 initial + 2 refinement) ✓')
print(f'  Final result has no error ✓')
print(f'  Refinement prompt contained _ERROR_SUFFIX with error message ✓')
snippet_start = error_prompt.find("EXECUTION ERROR")
print(f'\n  Refinement prompt snippet:\n    ...{error_prompt[snippet_start:snippet_start + 120]}...')
print('\n=== All Section 7 tests passed ✓ ===')
```
